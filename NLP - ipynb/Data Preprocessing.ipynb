```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
import pandas as pd
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer()
```

```python
spam=pd.read_csv("SPAM Data.csv")
```

```python
new_stopwords=stopwords.words("english")
```

```python
def datacleaning(x):
    y=nltk.word_tokenize(x)
    temp=[]
    for i in y:
        if i in new_stopwords:
            pass
        else:
            temp.append(i)
    string=' '.join(temp)
    return (string)
```

```python
spam['Clean_message']=spam['Message'].apply(datacleaning)
```

```python
spam
```

```python
def my_lemmatizer(a):
    c=nltk.word_tokenize(a)
    list=[]
    for i in c:
        list.append(lemmatizer.lemmatize(i))
        new_string=' '.join(list)
    return new_string
```

```python
spam['Lemmatized_message']=spam['Clean_message'].apply(my_lemmatizer)
```

```python
spam
```

```python
def Text_count(x):
    y=nltk.word_tokenize(x)
    temp=[]
    for i in y: 
         temp.append(i)
    d=' '.join(temp)
    return(len(d.split()))
```

```python
spam['Text_Length']=spam['Clean_message'].apply(Text_count)
```

```python
spam
```

```python
def remove_num(x):
    temp=[]
    for i in x:
        if i.isdigit(): 
            pass
        else:
            temp.append(i)
    my_newsttr=''.join(temp)
    return my_newsttr
```

```python
remove_num("i have 5 bars right now and 3 at home")
```

```python
spam['No_numbers']=spam['Clean_message'].apply(remove_num)
```

```python
spam
```

```python
vectorizer.fit(spam['No_numbers'])
```
